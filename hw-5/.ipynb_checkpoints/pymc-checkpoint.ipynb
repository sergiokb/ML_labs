# PyMC3



## 1. Создание модели

Модель определяется с помощью класса `Model` внутри контекста `with` 

```python
with pm.Model() as model:
    pass
```

Рассмотрим следующую модель:

$X = (X_1, ..., X_n) \sim \mathcal{N}(\mu, 1)$ &mdash; выборка

$\mu \sim \mathcal{N}(0, 1)$ &mdash; априорное распределение параметра

Зададим выборку:

```python
sample = sps.norm.rvs(size=100)
```

Теперь определяем модель

```python
with pm.Model() as model:
    mu = pm.Normal('mu', mu=0, sigma=1)
    X = pm.Normal('X', mu=mu, sigma=1, observed=sample)
```

В качестве `observed` поддерживаются списки `numpy.ndarray`, `theano` и `pandas` структуры.

Список всех случайных величин модели

```python
model.basic_RVs
```

Список ненаблюдаемых величин

```python
model.free_RVs
```

Список наблюдаемых величин

```python
model.observed_RVs
```

## 2. Распределения

Список всех вероятностных распределений
https://docs.pymc.io/api/distributions.html

Например, вот так можно задать нормальное распределение

```python
with pm.Model():
    x = pm.Normal('x', mu=0, sigma=1)
```

Матрица 3x3 из независимых случайных величин, имеющих бета-распределение

```python
with pm.Model():
    p = pm.Beta('p', 1, 1, shape=(3, 3))
```

Списки случайных величин тоже можно делать, но не рекомендуется

```python
with pm.Model():
    x = [pm.Normal('x_{}'.format(i), mu=0, sigma=1) for i in range(10)] # bad
```

Детерминированные операции над случайными величинами

```python
with pm.Model():
    x = pm.Normal('x', mu=0, sigma=1)
    y = pm.Gamma('y', alpha=1, beta=1)
    plus_2 = x + 2
    summed = x + y
    squared = x**2
    sined = pm.math.sin(x)
```

Результаты таких операций не сохраняются автоматически. Чтобы их отслеживать, стоит использовать `pm.Deterministic`:

```python
with pm.Model():
    x = pm.Normal('x', mu=0, sigma=1)
    plus_2 = pm.Deterministic('x plus 2', x + 2)
```

## 3. Семплрование

Работа методов семплироания производится с помощью функции `pm.sample()`.
По умолчанию эта функция пытается автоматически назначить сэмплер и инициализацию.


Сгенерируем выборку размера 1000 из апостериорного распределения и позволим семплеру настроиться на дополнительных `tune=500` итерациях, которые отбрасываются. Параллельно будет запущенно `cores=4` независимых цепи.

```python
sample = sps.norm.rvs(size=100)

with pm.Model() as model:
    mu = pm.Normal('mu', mu=0, sigma=1)
    sd = pm.HalfNormal('sd', sigma=1)
    x = pm.Normal('x', mu=mu, sigma=sd, observed=sample)

    trace = pm.sample(1000, tune=500, cores=4)
```

Как видим, для непрерывной модели PyMC3 назначает семплер NUTS, который очень эффективен даже для сложных моделей.

Извлекаем результаты генерации

```python
trace
```

```python
trace.posterior.mu
```

Наконец, сами значения

```python
trace.posterior.mu.values
```

Сгенерировано 4000 точек &mdash; по 1000 каждой из 4-х цепей

```python
trace.posterior.mu.values.shape
```

Для анализа результатов семплирования выборки часто используют так называемый trace-plot. Для каждого параметра рисуется ядерная оценка плотности (или гистограмма для дискретных) и траектория марковской цепи. Каждой независимой цепи соответствует свой цвет. Отрисовка результатов для каждой цепи отдельно полезна для контроля сходимости метода. На графиках траекторий полезно смотреть, действительно ли траекторию можно считать выборкой, или же в ней есть сильные автокорреляции.

```python
pm.plot_trace(trace, figsize=(10, 4));
```

Можно посмотреть также на `forestplot` &mdash; интервалы по каждой из 4-х марковских цепей

```python
pm.plot_forest(trace, linewidth=3, hdi_prob=0.95);
```

Наконец, можно извлечь полезные свойства апостериорного распределения

```python
pm.plot_posterior(trace, linewidth=3, hdi_prob=0.95, figsize=(10, 4));
```

### 4. Байесовская регрессия

Теперь попробуем применить это все к машинному обучению. Создадим линейные данные:

$Y = \theta_0 + \theta_1 X + \varepsilon$

и будем обучать линейную регрессию.


Создадим данные.

```python
N = 100 # размер выборки
X = np.random.uniform(low=0, high=100, size=N)
Y = 3 - 5 * X + np.random.normal(size=N, scale=30) 
```

```python
sns.scatterplot(x=X, y=Y)
plt.title("Искусственные данные для регрессии")
plt.xlabel("Признак")
plt.ylabel("Таргет")
plt.show()
```

Создадим модель. 

Зададим параметры и распределения данных:
$$\theta_0, \theta_1 \sim \mathcal{N}(0, 20)$$
$$\sigma^2 \sim \mathcal{HalfCauchy}$$
$$Y = \theta_0 + \theta_1 X + \varepsilon, \text{ где } \varepsilon \sim \mathcal N(0, \sigma^2 I_n)$$

*Замечание:* для того, чтобы связать признак и таргет и иметь возможность далее предиктивно семплировать, нам понадобится опция `dims`. Его надо указать в `X` и `Y`, чтобы потом мы могли семпировать данные длины, не совпадающей с обучающей выборкой.

```python
with pm.Model() as model:
    # Априорные распределения
    
    #параметр распределения eps
    sigma_prior = pm.HalfCauchy('sigma', beta=10, initval=1.) 
    # коэфициенты модели
    theta_0_prior = pm.Normal('theta_0', 0, sigma=20)
    theta_1_prior = pm.Normal('theta_1', 0, sigma=20)
    
    # объявляем данные
    x = pm.MutableData('x', X, dims='dims') # подаем обучающую выборку (X)

    # распределение данных
    y_distr = pm.Normal(
        'y', 
        # предположение, что y = theta0 + theta1 * x + eps
        mu=theta_0_prior + theta_1_prior * x, # theta0 + theta1 * x, центр распределения
        sigma=sigma_prior, # появляется из-за того, что добавляем eps
        observed=Y, # подаем обучающую выборку (Y)
        dims='dims' # связываем с x-ом
    )

    trace = pm.sample(3000, chains=2)
```

```python
pm.plot_trace(trace, figsize=(10, 5))
plt.tight_layout()
```

Моды распределений расположены довольно близко к реальным значениям.


Построим теперь предсказательный интервал, взяв область, в которую попадает 95% вероятностной массы апостериорного распределения. 

```python
x_grid = np.linspace(0, 100, 1000) 
# можем подать сюда тестовую выборку вместо простой сетки
# тогда можем семплировать предсказания
with model:
    pm.set_data({'x': x_grid}) # задаем новые X
    # вот это бы не работало без dims:
    post_pred = pm.sample_posterior_predictive(trace) # предсказываем
```

```python
post_pred.posterior_predictive['y'].values.shape 
# chains X iter X len(x_grid)
```

```python
# возьмем значения для chain=0
y_pred = post_pred.posterior_predictive['y'].values[0] 
# 95%-й доверительный интервал, будет размера 2 x len(x_grid)
bounds = np.percentile(y_pred, q=[2.5, 97.5], axis=0) 
# берем перценили так, чтобы между ними было 95% вероятностной массы
# bounds[0] - массив нижних границ, bounds[1] - верхних

sns.set(font_scale=1.5, style='whitegrid', palette='Set2')
plt.figure(figsize=(8, 7))
sns.scatterplot(x=X, y=Y, alpha=0.5)
plt.fill_between(x_grid, bounds[0], bounds[1], alpha=0.2, 
                 color='green', label='предсказательный интервал')
plt.xlabel("Признак")
plt.ylabel("Таргет")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
