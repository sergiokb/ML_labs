# 1. Байесовский подход
__________________

Рассмотрим схему испытаний Бернулли (т.е. броски монет) с вероятностью успеха $p$. 

**1).** Построим несколько графиков сопряженного распределения для разных параметров.
Можно рассмотреть следующие случаи.
* Монета скорее честная. Тогда наиболее вероятны значения $p$ в окрестности $0.5$.
* Монета скорее нечестная, перевес неизвестен. Тогда наименее вероятны значения $p$ в окрестности $0.5$.
* Монета скорее нечестная, перевес в сторону герба. Тогда наиболее вероятны значения $p$ в окрестности $1$.
* Монета скорее честная, либо с небольшим перекосом вправо. Тогда наиболее вероятны значения $p$ в окрестности $\approx 0.6$.
* Ничего не известно. Тогда все значния равновероятны.


Параметры апостериорного распределения $Beta(\alpha, \beta)$ можно интерпретировать следующим образом:

* $\alpha$ &mdash; перевес в сторону герба

* $\beta$ &mdash; перевес в сторону решки

* $\alpha + \beta$ &mdash; эффективный размер выборки, то есть чем больше мы знаем про параметр $\theta$ распределения $Bern(\theta)$, тем больше сумма $\alpha + \beta$

```python
param_list = [[5, 5], [1/2, 1/2], [5, 1], [6, 4], [1, 1]]
label_list = ["Честная", 
              "Нечестная, перевес неизвестен", 
              "Нечестная, перевес в сторону герба", 
              "Честная, либо с небольшим \nперекосом в сторону герба", 
              "Ничего не известно"]
grid = np.arange(0, 1, 0.01)

plt.figure(figsize=(12, 6))
for param, label in zip(param_list, label_list):
    plt.plot(grid, sps.beta(a=param[0], b=param[1]).pdf(grid), label=label, lw=5)
    plt.xlabel("Вероятность успеха")
plt.title("Монета")
plt.legend()
plt.show()
```

**Вывод:** на графиках выше визуализировали выборку из распределения $Bern(\theta)$ с помощью бета распределения. Из данного графика можно предположить как именно были получены данные, то есть оценить значение параметра $\theta$.

Ниже приведена реализация некоторых вспомогательных функций.

```python
def draw_posteriori(grid, distr_class, post_params, xlim=None):
    '''Рисует серию графиков апостериорных плотностей.
    
    :param grid: сетка для построения графика
    :param distr_class: класс распределений из scipy.stats
    :param post_params: параметры апостериорных распределений 
        shape=(размер выборки, кол-во параметров)
    '''
    
    sns.set(font_scale=1.3)
    
    size = post_params.shape[0] - 1
    
    plt.figure(figsize=(12, 5))
    for n in range(size+1):
        plt.plot(grid, 
                 distr_class(post_params[n]).pdf(grid) \
                     if np.isscalar(post_params[n]) \
                     else distr_class(*post_params[n]).pdf(grid), 
                 label='n={}: {}'.format(n, np.round(post_params[n], 3)), 
                 lw=2.5,
                 color=(1-n/size, n/size, 0))
    plt.title('Апостериорное распределение в зависимости от размера выборки')
    # plt.grid(ls=':')
    plt.legend(fontsize=12, loc=2)
    plt.xlim(xlim)
    plt.show()
    

def draw_estimations(ml, distr_class, post_params, confint=True, ylim=None):
    '''Рисует графики байесовской оценки (м.о. и дов. инт.) и ОМП.
    
    :param ml: оценка максимального правдоподобия для 1 <= n <= len(sample)
    :param distr_class: класс распределений из scipy.stats
    :param post_params: параметры апостериорных распределений 
        shape=(размер выборки, кол-во параметров)
    '''
    
    sns.set(font_scale=1.3, style='whitegrid', palette='Set2')
    
    size = len(ml)
    distrs = []
    for n in range(size+1): 
        distrs.append(distr_class(post_params[n]) \
                          if np.isscalar(post_params[n]) \
                          else distr_class(*post_params[n]))
    
    plt.figure(figsize=(12, 3))
    plt.plot(np.arange(size+1), [d.mean() for d in distrs], 
             label='Bayes', lw=2.5)
    plt.fill_between(np.arange(size+1), [d.ppf(0.975) for d in distrs], 
                     [d.ppf(0.025) for d in distrs], alpha=0.3)
    plt.plot(np.arange(size)+1, ml, label='MLE', lw=2.5)
    plt.title('Оценки параметра в зависимости от размера выборки')
    plt.xlabel('Размер выборки')
    # plt.grid(ls=':')
    plt.ylim(ylim)
    plt.legend()
    plt.show()
```

```python
def bern_posterior_params(sample, a, b):
    '''Находит параметры апостериорного распределения 
        для всех 0 <= n <= len(sample). 
    
    :param sample: выборка
    :param a: параметр alpha априорного распределения
    :param b: параметр beta априорного распределения
    
    :returns: параметры апостериорного
    '''
    
    n = np.arange(sample.shape[0]) + 1
    params = list()
    params.append(a + sample.cumsum())
    params.append(b + n - sample.cumsum())
    params = np.array(params).T
    return params
```

```python
def MLE(sample):
    n = np.arange(sample.shape[0]) + 1
    ml = np.cumsum(sample)/n
    return ml[1:]
```

Сгенерируем по 15 бросков симметричной и несимметричной монет и рассмотрим для каждой из них разные случаи выбора параметров априорного распределения

```python
n = 15

p_fair = 0.5
p_unfair = 0.9

fair_sample = sps.bernoulli(p=p_fair).rvs(size=n)
unfair_sample = sps.bernoulli(p=p_unfair).rvs(size=n)
grid = np.arange(0, 1, 0.005)
```

```python
coin_is_fair_list = [1, 1, 0, 0]
prior_params = [[5, 5], [7, 1], [7, 1], [5, 5]]

for (a, b), coin_is_fair in zip(prior_params, coin_is_fair_list):

    title = (
        f"Монета: p={p_fair if coin_is_fair else p_unfair}. "
        f"Параметры априорного: "
        f"a={a}, b={b}")
        
    sample = fair_sample if coin_is_fair else unfair_sample
        
    post_params = bern_posterior_params(sample, a, b)

    # print('---------------\n')
    # print(title)
    draw_posteriori(grid, sps.beta, post_params, xlim=(0, 1))
    draw_estimations(MLE(sample), sps.beta, post_params, confint=True, ylim=None)
```

**Вывод:** 

* при маленьком размере выборки апостериорное распределение практически не отличается от априорного, поэтому байесовская оценка сильно зависит от наших начальных предположений


* с ростом размера выборки пик апостериорного распределения становится более выраженным: плотность распределения начинает концентрироваться рядом с истинным значением параметра, основываясь на данных эксперимента. Аналогичный эффект можно наблюдать при визуализации доверительного интервала: он сужается с ростом размера выборки. 


* по ЗБЧ оценка максимального правдоподобия стремится к истинному значению параметра с ростом размера выборки


# 2. Метод Монте-Карло

### 2.1 Вычисления мат. ожиданий

Рассмотрим случайную величину $\xi$, равномерно распределенную на $[0, \pi]$, и, для примера функцию $f(x) = x^2 - 10x + 2 \sin x$. Подсчитаем математическое ожидание $\mathsf E f(\xi)$ с помощью метода Монте-Карло и сравним с аналитически вычисленным значением.

```python
def f(x): # реализация функции
    return x ** 2 - 10 * x + 2 * np.sin(x)
```

```python
%%time
N = 10_000_000 # размер выборки
X = np.random.uniform(low=0, high=np.pi, size=N)
print(f(X).mean())
```

Можно подсчитать, что мат.ожидание равно $\frac 4 \pi - 5 \pi + \frac {\pi ^ 2} 3 \approx -11.1448555895$. Таким образом, мы вычислили искомое значение менее чем за секунду и довольно точно, не прибегая к мат. анализу. Ура!

Таким образом, можно применять этот метод к вычислению трудно берущихся или неберущихся интегралов.

----

### 2.2 Численное интегрирование

Реализуем метод Монте-Карло численного интегрирования функции. Будем вычислять интеграл вида
$$ \int\limits_{l_1}^{h_1} ... \int\limits_{l_d}^{h_d} f(x_1, ... , x_d) dx_1 ... dx_d $$

```python
def integrate_monte_carlo(f, low, high, n):
    """
    Метод монте-карло
    f --- векторная функция d-мерной переменной.
        Реализация функции на вход принимает матрицу (n, d),
        где n - число точек, в которых нужно вычислить функцию,
        а d - размерность пространства переменных.
        Функция должна вернуть вектор размерности n.
    low, high --- нижние и верхние пределы интегрирования.
        Передаются в том же порядке, в котором функция 
        принимает аргументы).
    n --- максимальное число вызовов функции f.
    """
    low = np.asarray(low)
    high = np.asarray(high)
    dim = len(high)
    # случайные точки, в которых будем вычислять функцию
    dots = sps.uniform.rvs(size=(n, dim), loc=low, scale=high-low)  

    # вычисление функции в случайных точках
    return (np.sum(f(dots)) * np.prod(high - low)) / n
```

Вычислим $\int\limits_0^1 \int\limits_3^4 \left(x^2 + y^2\right) dxdy$ на миллионе запусков функции $f(x, y) = x^2+y^2$ двумя рассмотренными методами.

Инициализируем переменные и функцию

```python
n = 1_000_000 # объявляю константы для передачи в функцию 
low = [3, 0] # нижние пределы интегрирования
high = [4, 1] # верхние

square_sum = lambda x: np.sum(x**2, axis=1)
```

Вычислим интеграл по методу Монте-Карло

```python
%%time
mn = integrate_monte_carlo(square_sum, low, high, n)
mn
```

Если посчитать аналитически, чему равен интеграл, то получится $12 \frac 2 3 $, то есть мы получили достаточно неплохое приближение.

Посмотрим теперь на скорость сходимости

```python
max_dim = 2

def plot_convergence(f, low, high, n_list, theoretical_value, max_dim):
    """
    Построение графиков сходимости методов Монте-Карло
    
    Аргументы:
        f --- векторная функция d-мерной переменной.
            Реализация функции на вход принимает матрицу (n, d),
            где n - число точек, в которых нужно вычислить функцию,
            а d - размерность пространства переменных.
            Функция должна вернуть вектор размерности n.
        low, high --- нижние и верхние пределы интегрирования.
            Передаются в том же порядке, в котором функция 
            принимает аргументы).
    n_list --- список.
    theoretical_value --- точное значение интеграла.
    """
    monte_carlo_computed = np.zeros(len(n_list))

    for n in np.arange(len(n_list)):
        mn = integrate_monte_carlo(f, low, high, n_list[n])
        monte_carlo_computed[n] = mn

    plt.figure(figsize=(9, 5))
    plt.title('Сходимость метода Монте-Карло')      
    plt.xlabel("Число вызовов функции")
    plt.ylabel("Значение интеграла")
    plt.plot(n_list, monte_carlo_computed, label="Оценка")
    plt.axhline(y=theoretical_value, color='indigo', linestyle='--', label="Истинное значение")
    plt.legend(fontsize=14)
```

```python
theoretical_value = 2 / 3

n_list = np.arange(1000, 50001, 100)
low = np.array([0, 0])
high = np.array([1, 1])

plot_convergence(square_sum, low, high, n_list, 
                 theoretical_value, max_dim)
```

### 2.3 Применение метода в Байесовской статистике

Пусть $X_1, ..., X_n$ &mdash; выборка из условного распределения $\mathcal{N}(\theta, \sigma^2)$, $\theta \in \Theta = \mathbb{R}$ и $\theta$ в качестве априорного распределения имеет стандартное распределение Коши со сдвигом. 

Апостериорное распределение вычисляется по формуле:
$$q\left.(t\,\right|\,x) = \frac {q(t) p_t(x_1) ... p_t(x_n)} {\int\limits_\Theta q(t) p_t(x_1) ... p_t(x_n) dt},$$ 
где $p_t(x)$ &mdash; плотность распределения $\mathcal{N}(t, \sigma^2)$, а $q(t)$ &mdash; плотность распределения Коши.

---

Как было сказано на лекции, аналитически интеграл в знаменателе посчитать не удастся. Однако, этот интеграл можно вычислить численно с помощью метода Монте-Карло.

Рассмотрим столбец Alcalinity of ash датасета <a href="http://archive.ics.uci.edu/ml/datasets/Wine">о вине</a>.


```python
names = """\
0) Target
1) Alcohol
2) Malic acid
3) Ash
4) Alcalinity of ash
5) Magnesium
6) Total phenols
7) Flavanoids
8) Nonflavanoid phenols
9) Proanthocyanins
10)Color intensity
11)Hue
12)OD280/OD315 of diluted wines
13)Proline"""

names = [x.split(")")[1].strip() for x in names.split("\n")]
```

```python
df = pd.read_csv("./wine/wine.data", header=None)
df.columns = names
df.head()
```

Посмотрим на гистограмму выбранного признака.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Alcalinity of ash"], stat='density', alpha=0.9)
sns.kdeplot(df["Alcalinity of ash"], linewidth=3, color='green', alpha=0.9)
plt.show()
```

Наше предпложение о его распределении похоже на правду.

Параметры априорного распределения подбираем исходя из статистических данных, найденных в интернете, например <a href="https://www.institut-heidger.de/ash-gravimetric-calculated-from-minerals-alkalinity-of-ash/?lang=en">здесь</a>. Так как величина принимает значение от 13 до 35 единиц, выберем среднее распределения коши в 24 единицы, что недалеко от выборочного среднего, а дисперсию $\sigma = \frac{35 - 24}{3} =  \frac{24 - 13}{3} = \frac{11}{3}$, подгоняя интервал под правило трёх сигм.

```python
sigma = 11./3  # выбранное сигма
sample_norm = df["Alcalinity of ash"].values  # выборка из нормального
```

Посчитаем исходный интеграл методом Монте-Карло.

```python
def eval_c(sample_norm, cauchy_sample_size=100000):
    '''
    Функция для оценки константы в знаменателе формулы Байеса.
    
    :param sample_norm: выборка из нормального распределения.
    :param cauchy_sample_size: размер выборки для метода Монте-Карло
                        
    :returns: оценка константы
    '''
    
    sample_cauchy = sps.cauchy(loc=24).rvs(
        size=cauchy_sample_size
    ).reshape(1, -1)
    sample_norm = sample_norm.reshape(-1, 1)
    
    return np.mean(
        sps.norm(loc=sample_cauchy, scale=sigma).pdf(sample_norm).prod(axis=0)
    )
```

Посчитаем значение константы $c$ &mdash; знаменателя в формуле Байеса. Оно получается довольно маленьким

```python
c = eval_c(sample_norm, 100000)
print(f'1/c = {1./c:.2e}')
print(f'c = {c:.2e}')
```

Теперь, согласно формуле, строим апостериорную плотность:

```python
# сетка и подсчет плотности
grid = np.linspace(-3 + sample_norm.mean(), 3 + sample_norm.mean(), 500000)

density = np.exp(-((sample_norm.reshape(-1, 1) - \
                    grid) / sigma)**2 / 2).prod(axis=0) * \
                    sps.cauchy(loc=24.).pdf(grid) / c * \
                    np.power(1 / np.sqrt(2 * np.pi) / sigma, len(sample_norm))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10, 5))
plt.title('График апостериорной плотности')

plt.plot(grid, density, label=r"$q(t | x)$", lw=4)
plt.xlabel("Параметр $\\theta$")
plt.legend()
plt.show()     
```

Вычислим матожидание:

```python
def eval_mean(c, sample_norm, cauchy_sample_size=500000):
    '''
    Функция для подсчет математического ожидания апостериорного распределения.
    
    :param c: консанта в формуле Байеса
    :param sample_norm: выборка из нормального распределения
    :param cauchy_sample_size: размер выборки для метода Монте-Карло
    
    :returns: посчитанное математическое ожидание
    
    '''
    
    sample_cauchy = sps.cauchy(loc=24).rvs(size=cauchy_sample_size).reshape(1, -1)
    sample_norm = sample_norm.reshape(-1, 1)
    to_return = np.mean([
        sample_cauchy * \
        np.exp(-((sample_norm - sample_cauchy) / sigma)**2 / 2).prod(axis=0) \
        / c * np.power(1 / np.sqrt(2 * np.pi) / sigma, len(sample_norm))
    ])
    return to_return
```

```python
c = eval_c(sample_norm, 500000)
mean_evaluated = eval_mean(c, sample_norm)
mean_evaluated
```

Найдём доверительный интервал. Для этого численно вычислим функцию распределения, зная плотность и постоянный шаг сетки точек, в которых плотность вычислена:

```python
step = grid[1] - grid[0]
F = density.cumsum() * step
alpha = 0.05
bounds = grid[F >= alpha / 2][0], grid[F <= 1 - alpha /2][-1]
F_in_left_bound, F_in_right_bound = F[F >= alpha / 2][0], F[F <= 1 - alpha /2][-1]
```

Границы доверительного интервала:

```python
bounds
```

Вероятность попасть в этот интервал:

```python
F_in_right_bound - F_in_left_bound
```

Метод Монте-Карло позволил нам численно посчитать константу в знаменателе формулы Байеса, а также математическое ожидание с хорошой точностью. После вычисления $c$ мы смогли построить график апостериорной плотности найти 95% доверительный интервал.
