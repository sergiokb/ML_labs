### Задача 1.

Ученые придумали новый тест для определения одной очень известной болезни, который показывает поразительные свойства:
* Вероятность того, что тест покажет положительный результат, если человек и правда болен, составляет 99%;
* Вероятность того, что тест покажет положительный результат, если человек на самом деле не болен, составляет 3%.

Посчитайте вероятность того, что человек реально болен, если тест показал положительный результат, с помощью формулы Байеса. Также известно, что в среднем на 100 случайных человек приходится двое болеющих.


*Пусть $A$ - событие того, что тест показал положительный результат, а $B$ - событие того, что человек болен. \
Тогда вероятность того, что человек реально болен, если тест показал положительный результат:* \
$P(B|A) = \dfrac{P(A|B)P(B)}{P(A)} = \dfrac{P(A|B)P(B)}{P(A|B)P(B) + P(A|\overline{B})P(\overline{B})}$ \
*При этом известно, что:* 
* $P(A|B) = 0.99$
* $P(A|\overline{B}) = 0.03$
* $P(B) = 0.02 ⇒ P(\overline{B}) = 0.98$

*Таким образом,* \
$P(B|A) = \dfrac{0.99 \cdot 0.02}{0.99 \cdot 0.02 + 0.03 \cdot 0.98} = \dfrac{198}{492} = \dfrac{33}{82} ≈ 40\%$

### Задача 2.

Рассмотрим два случая.

**a).** Пусть $X_1, \dots, X_n$ &mdash; выборка из распределения $\mathcal{N}(\theta, 1)$.

**b).** Пусть $X_1, \dots, X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \theta^{-1})$.

В каждом случае подберите сопряженное семейство распределений и найдите соответствующее апостериорное распределение параметра $\theta$, а также его математическое ожидание.

Подумайте, какое распределение является сопряженным в случае если неизвестны оба параметра нормального распределения.


**a).** $X_1, \dots, X_n \sim \mathcal{N}(\theta, 1)$. \
$$ p_t(x) = \left(\frac{1}{2\pi}\right)^{\frac{n}2} e^{-\frac12\sum_{i = 1}^{n}(x_i - t)^2} = \left(\frac{1}{2\pi}\right)^{\frac{n}2} e^{-\frac12\sum_{i = 1}^{n}x_i^2 + (\sum_{i = 1}^{n}x_i)t-\frac{n}2t^2} $$ \
*Хотим:* $q(t|x) \propto q(t)p_t(x)$ \
*Априорное распределениe параматра $\theta$ зададим  как $\mathcal{N}(a, \sigma^2)$. Тогда* \
$$q(t) = \frac1{\sqrt{2\pi\sigma^2}}e^{-\frac1{2\sigma^2}(t - a)^2} =  \frac1{\sqrt{2\pi\sigma^2}}e^{-\frac1{2\sigma^2}t^2 +\frac{a}{\sigma^2}t-\frac1{2\sigma^2}a^2}$$ \
$$q(t|x) \propto q(t)p_t(x) \propto e^{-\frac12(\frac1{\sigma^2}+n)t^2 +(\frac{a}{\sigma^2}+\sum_{i = 1}^{n}x_i)t}$$
<!-- $$q(t|x) \propto \mathcal{N}\left(\frac1{\frac1{\sigma^2}+n}(\frac{a}{\sigma^2}+\sum_{i = 1}^{n}x_i), \frac1{\frac1{\sigma^2}+n}\right)$$ -->
$$q(t|x) \propto \mathcal{N}\left(\frac{(\frac{a}{\sigma^2}+\sum_{i = 1}^{n}x_i)}{\frac1{\sigma^2}+n}, \frac1{\frac1{\sigma^2}+n}\right)$$
$$\mathbb{E}(\theta|x) = \frac{(\frac{a}{\sigma^2}+\sum_{i = 1}^{n}x_i)}{\frac1{\sigma^2}+n}$$

**б).** $X_1, \dots, X_n \sim \mathcal{N}(0, \theta)$. \
$$ p_t(x) = \left(\frac{1}{2\pi t}\right)^{\frac{n}2} e^{-\frac1{2t}\sum_{i = 1}^{n}x_i^2}$$ \
*Хотим:* $q(t|x) \propto q(t)p_t(x)$ \
*Априорное распределениe параматра $\theta$ зададим  как обратное гамма распределение $\hat\Gamma(\alpha, \beta)$:* \
$$q(t) = \frac{\alpha^{\beta}t^{-1-\beta}}{\Gamma(\beta)}e^{-\frac{\alpha}{t}}$$
$$q(t|x) \propto q(t)p_t(x) \propto t^{-1-(\frac{n}2 +\beta)}e^{-\frac1{t}({\alpha} +\frac12\sum_{i = 1}^{n}x_i^2)}$$
$$q(t|x) \propto \hat\Gamma({\alpha} +\frac12\sum_{i = 1}^{n}x_i^2, \beta +\frac{n}2)$$
*Найдем матожидание обратного гамма-распределения $\hat\Gamma(\alpha, \beta)$:* \
$\mathbb{E}[t] = \int_{0}^{\infty} t q(t) dt = \int_{0}^{\infty} t \frac{\alpha^{\beta}t^{-1-\beta}}{\Gamma(\beta)}e^{-\frac{\alpha}{t}} dt = \int_{0}^{\infty} \frac{\alpha^{\beta} t^{-\beta+2} e^{-\frac{\alpha}{t}}}{\Gamma(\beta)}\frac1{t^2}dt = \int_{0}^{\infty} \frac{\alpha^{\beta} u^{\beta-2} e^{-\alpha u}}{\Gamma(\beta)}du = \frac{\alpha\Gamma(\beta - 1)}{\Gamma{(\beta)}}\int_{0}^{\infty} \frac{\alpha^{\beta-1} u^{\beta-2} e^{-\alpha u}}{\Gamma(\beta - 1)}du = \frac{\alpha\Gamma(\beta - 1)}{\Gamma{(\beta)}} = \frac{\alpha}{\beta-1}$ \
*Тогда*: $$\mathbb{E}(\theta|x) = \frac{{\alpha} +\frac12\sum_{i = 1}^{n}x_i^2}{\beta +\frac{n}2 - 1 }$$

### Задача 3.

Рассмотрим столбец Alcalinity of ash датасета <a href="http://archive.ics.uci.edu/ml/datasets/Wine">о вине</a>. Цель &mdash; оценить среднего значения этого признака.

Для простоты дальнейших вычислений выберите некоторым образом дисперсию данных и считайте ее известной. На практике это не так, и дисперсию можно и нужно оценивать, но мы не хотим сильно усложнять учебную задачу.

```python
import numpy as np
import scipy.stats as sps
import pandas as pd
import math

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2, palette='Set1')
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Alcalinity of ash"], stat='density', alpha=0.6)
sns.kdeplot(df["Alcalinity of ash"], linewidth=3, color='green', alpha=0.6)
plt.show()
```

Из класса сопряженных распределений выберите априорное распределение для среднего значения на основе своих представлений, не используя данные. Подсказка: загляните в семинар.

*Распределение на графике похоже на нормальное, поэтому будем считать, что распределено как* $\mathcal{N}(\theta, \sigma^2)$ \
*На основании данных из интернета, выберем априорную $\sigma = \frac{11}{3}$*

$X_1, \dots, X_n \sim \mathcal{N}(\theta, \sigma^2)$. \
*Априорное распределениe параматра $\theta$ зададим  как $\mathcal{N}(a, s^2)$. Тогда по аналогии с первой задачей* \
$$q(\theta|x) \sim \mathcal{N}\left(\frac{(\frac{a}{s^2}+\frac1{\sigma^2}\sum_{i = 1}^{n}x_i)}{\frac{n}{\sigma^2}+\frac1{s^2}}, \frac{1}{\frac{n}{\sigma^2}+\frac1{s^2}}\right)$$
$$\mathbb{E}(\theta|x) = \frac{(\frac{a}{s^2}+\frac1{\sigma^2}\sum_{i = 1}^{n}x_i)}{\frac{n}{\sigma^2}+\frac1{s^2}}$$

```python
sample_norm = df["Alcalinity of ash"].values # выборка из нормального
sigma = 11./3  # выбранное сигма 
a0, s0 = 24, 4
```

```python
def recount(sigma, a, s, x_es):
    a_new = (a / (s ** 2) + np.sum(x_es) / (sigma ** 2)) / (len(x_es) / (sigma ** 2) + 1 / (s ** 2))
    s_new = 1 / (len(x_es) / (sigma ** 2) + 1 / (s ** 2))
    return a_new, s_new
```

Визуализируйте динамику изменения апостериорного распределения, последовательно добавляя к данным по 20 объектов. Примеры кода можно посмотреть в семинаре.

```python
a, s = a0, s0
plt.figure(figsize=(12, 6))
grid = np.linspace(10, 30, 100000)
plt.plot(grid, sps.norm(a, np.sqrt(s)).pdf(grid), label=f'size=0')
for size in 20 * (np.arange(len(sample_norm) // 20) + 1):
    a, s = recount(sigma, a, np.sqrt(s), sample_norm[size-20:size])
    plt.plot(grid, sps.norm(a, np.sqrt(s)).pdf(grid), label=f'size={size}')
plt.xlim(12, 30)
plt.legend()
plt.show()
print("Last expected value:", a)
print("Last variety:", s)
```

**Вывод:** *Таким образом, байесовский подход работает великолепно! С каждым добавленным куском выборки график вытягивается, а значит повышается уверенность оценки.*

### Задача 4.

Некоторая ML-модель имеет один параметр $\theta$, который обучается посредством ***минимизации*** функции
$$\mathcal{L}(\theta) = \mathsf{E} \exp\left(-\frac{\xi^2\sqrt{\eta}}{1+\theta^2}\right),$$
где $\xi$ имеет стандартное нормальное распределение, а $\eta$ &mdash; пуассоновское распределение с параметром 5 и не зависит от $\eta$.


Оптимизируйте эту функцию, используя стохастический градиентный спуск. Необходимо записать формулу шага и реализовать процедуру оптимизации.

<details>
  <summary>Подсказка</summary>

Посчитайте производную этой функции по $\theta$, поменяв местами интеграл (мат. ожидание) и производную. Далее оцените мат. ожидание по методу Монте-Карло.

</details>

$$\mathcal{L}'(\theta) = \mathsf{E} \frac{\partial}{\partial \theta}\exp\left(-\frac{\xi^2\sqrt{\eta}}{1+\theta^2}\right) = \mathsf{E} \left[\xi^2\sqrt{\eta}e^{-\frac{\xi^2\sqrt{\eta}}{1+\theta^2}}\frac{\partial}{\partial \theta}(\frac{-1}{1+\theta^2})\right] = \mathsf{E} \left[\frac{2\theta\xi^2\sqrt{\eta}}{(1+\theta^2)^2}e^{-\frac{\xi^2\sqrt{\eta}}{1+\theta^2}}\right] =$$ 
$$=\sum_{k=0}^{+\infty} p(\eta = k)\int\limits_{-\infty}^{+\infty}\frac{2\theta x^2\sqrt{\eta}}{(1+\theta^2)^2}e^{-\frac{x^2\sqrt{\eta}}{1+\theta^2}}p_{\xi}(x)dx \approx \sum_{k=0}^{m} p(\eta = k)\int\limits_{-l}^{l}\frac{2\theta x^2\sqrt{\eta}}{(1+\theta^2)^2}e^{-\frac{x^2\sqrt{\eta}}{1+\theta^2}}p_{\xi}(x)dx $$

*Воспользуемся методом Монте-Карло, подобрав значения $m$ и $l$:*

```python
m = sps.poisson(5).ppf(0.95)
l = sps.norm(0, 1).ppf(0.975)
m, l
```

```python
def underneath(theta, x, eta):
    return 2 * theta * (x ** 2) * np.sqrt(eta)/ (1 + theta ** 2) ** 2 * np.exp(-(x ** 2) * np.sqrt(eta) / (1 + theta ** 2)) * sps.norm(0, 1).pdf(x)
    
def get_grad(theta, size=1000):
    x_es = sps.norm.rvs(0, 1, size)
    etas = np.arange(m + 1)
    etas_expanded = np.expand_dims(etas, 1)
    integ = 2 * l * np.sum(underneath(theta, x_es, etas_expanded), axis=1) / size
    grad = np.sum(sps.poisson(5).pmf(etas) * integ)
    return grad
    
def regression(theta0=3, max_steps=5000, lr=0.01):
    theta = theta0
    thetas =[]
    for i in range(max_steps):
        thetas.append(theta)
        theta -= lr * get_grad(theta)
    # return theta
    plt.figure(figsize=(12, 3))
    plt.plot(np.arange(max_steps), thetas)
    plt.xlabel('Iteration')
    plt.ylabel(r'Value of $\theta$')
    plt.show()

regression()
```

*Таким образом, $\theta$ стремится к 0*
