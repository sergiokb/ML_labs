### Задача 1. 

> *В условии этой задачи подробно расписаны алгоритмы решения каждого пункта. Перед выполнением ознакомьтесь также с примерами, которые были рассказаны на лекции, они во многом аналогичны. При возникновении затруднений спрашивайте вопросы в чате.*

Пусть $X_1,...,X_n$ &mdash; выборка из экспоненциального распределения с плотностью $p_\theta(x) = \theta e^{-\theta x},$ где $x>0, \theta>0$. Такое распределение описывает время до наступления некоторого события. Например, таким событием может быть поступление очередного заказа таксисту или курьеру службы доставки. В таком случае $X_i$ &mdash; время ожидания $i$-го заказа после завершения предыдущего. Параметр $\theta$ имеет смысл интенсивности заказов в единицу времени. Следовательно, чем больше $\theta$, тем меньше среднее значение $X_i$. Точнее $\mathsf{E}_\theta X_i = 1/\theta$.


При выполнении задач используйте то, что $\mathsf{E}_\theta X_i^k = k!/\theta^k$.

**1.** Предложите асимптотически нормальную оценку $\theta$ и найдите ее асимптотическую дисперсию.

* Для выполнения этого пункта запишите центральную предельную теорему. Для какой функции $h(\theta)$ тем самым получилась асимптотически нормальная оценка?

* Какую функцию нужно применить, чтобы получить оценку для $\theta$?

* Воспользуйтесь дельта-методом, чтобы получить асимптотически нормальную оценку $\theta$. Не забудьте проверить применимость дельта-метода.

* Посчитайте ее асимптотическую дисперсию по формуле, которую дает дельта-метод. Обратите внимание, что в качестве аргумента производной нужно взять ту величину, для которой мы сначала построили оценку, то есть для $h(\theta)$.

**ЦПТ**: $\sqrt{n}(\overline{x} - a) \rightarrow \mathcal{N}(0, \sigma^2)$, где $a, \sigma^2$ - матожидание и дисперсия $x$ соответственно.  
В случае экспоненциального распределения:  $\sqrt{n}(\overline{x} - \dfrac1{\theta}) \rightarrow \mathcal{N}(0, \dfrac1{\theta^2})$, то есть асимптотически нормальная оценка получилась для функции $h(\theta) = \dfrac1{\theta}$  
Чтобы получить оценку для $\theta$, нужно применить функцию $g(y) = \dfrac1y,\ g'(y) = -\dfrac1{y^2}$  
**Дельта-метод:** $\sqrt{n}(g(\overline{x}) - g(a)) \rightarrow \mathcal{N}(0, (g'(a))^2\sigma^2)$  
$\sqrt{n}(\dfrac1{\overline{x}} - \dfrac1a)) \rightarrow \mathcal{N}(0, \dfrac1{a^4}\sigma^2)$  
$\sqrt{n}(\dfrac1{\overline{x}} - \theta)) \rightarrow \mathcal{N}(0, \theta^2)$  
**Итак, получили для $\theta$ асимптотически нормальную оценку $\theta^* = \dfrac1{\overline{x}}$ c асимптотической дисперсией $\theta^2$**

Сделайте это и получите еще одну асмптотически нормальную оценку $\theta$.

$\sqrt{n}\left(\sqrt{\dfrac2{\overline{x^2}}} - \theta\right) \rightarrow \mathcal{N}(0, \dfrac54\theta^2)$  
**Итак, получили для $\theta$ асимптотически нормальную оценку $\theta^{**} = \sqrt{\dfrac2{\overline{x^2}}}$ c асимптотической дисперсией $\dfrac54\theta^2$**

### Задача 2. 

В прошлой задаче вы получили два асимптотических доверительных интервала и один точный, а также соответствующие им статистические критерии. Самое время их сравнить на практике!

Число $\alpha$ всегда необходимо фиксировать до экспериментов, и ни в коем случае не подбирать по данным. Например, в случае AB-тестов это число может быть фиксировано отделом по рискам для всех экспериментов компании. В рамках задачи (и обычно на практике) мы используем уровень доверия $\alpha=0.95$, соответственно, уровень значимости критерия равен $1-\alpha=0.05$.

```python
import numpy as np
import pandas as pd
import scipy.stats as sps
from tqdm import tqdm
from functools import partial
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', palette='Set2', font_scale=1.6)
```

```python
def draw_confidence_interval(
    left,
    right,
    estimation=None,
    sample=None,
    ylim=(None, None),
    color_estimation='#FF3300',
    color_interval='#00CC66',
    color_sample='#0066FF',
    label_estimation='',
    label_interval='',
    label_element=None
):

    time = np.arange(len(left)) + 1
    assert len(time) == len(right)
    if not sample is None:
        plt.scatter(time, sample, alpha=0.3, color=color_sample, s=40, label=label_element)
        
    if not estimation is None:
        plt.plot(time, estimation, color=color_estimation, linewidth=2.5, label=label_estimation)

    plt.fill_between(time, left, right, alpha=0.25, color=color_interval, label=label_interval)
    plt.legend()
    plt.ylim(ylim)
    plt.xlabel('Размер выборки')
    plt.title('Зависимость доверительного интервала от размера выборки')

def draw_mean_lens(
    left,
    right,
    color='black',
):
    time = np.arange(len(left)) + 1
    mean_lens = np.cumsum(right - left) / time
    plt.plot(time, mean_lens, color=color, linewidth=3)
    plt.xlabel('Размер выборки')
    plt.title('Зависимость cредней длины доверительного интервала от размера выборки')
```

```python
def calculate_confidence_interval(
    sample,
    type,
    alpha=0.95,
    alpha1=0.025,
    return_estimation=False,
    cumulative=False,
    axis=-1,
):
    if cumulative:
        sums = np.cumsum(sample, axis=axis)
        sums_sq = np.cumsum(sample**2, axis=axis)
        size = np.arange(sample.shape[axis]) + 1
    else:
        sums = np.sum(sample, axis=axis)
        sums_sq = np.sum(sample**2, axis=axis)
        size = sample.shape[axis]
    mean = sums / size
    mean_sq = sums_sq / size
    if type == 'asymptotic1' or type == 'asymptotic2':
        quantile = sps.norm.ppf((1 + alpha) / 2)
        if type == 'asymptotic1':
            estimation = 1/mean
            left = estimation * (1 - 1/np.sqrt(size)*quantile)
            right = estimation * (1 + 1/np.sqrt(size)*quantile)
        if type == 'asymptotic2':
            estimation = np.sqrt(2/mean_sq)
            left = estimation * (1 - np.sqrt(5/(4*size))*quantile) 
            right = estimation * (1 + np.sqrt(5/(4*size))*quantile) 

    elif type == 'exact':
        estimation = mean
        quantile1 = sps.gamma.ppf(q=alpha1, a=size, scale=1/size)
        quantile2 = sps.gamma.ppf(q=alpha1+alpha, a=size, scale=1/size)
        left = quantile1 / mean
        right = quantile2 / mean
    
    if return_estimation:
        return left, right, estimation
    else:
        return left, right
```

```python
class Interval_relations():   
    def __init__(self, max_sample_size=30, sample=None, ylim=None, cumulative=True):
        self.theta = theta
        self.max_sample_size = max_sample_size
        self.sample = sample
        self.ylim = ylim
        self.cumulative = cumulative
        
    def count_it_all(self):
        self.left_asymptotic1, self.right_asymptotic1, self.estimation1 = calculate_confidence_interval(
            self.sample,
            type='asymptotic1',
            return_estimation=True,
            cumulative=self.cumulative
        )
        
        self.left_asymptotic2, self.right_asymptotic2, self.estimation2 = calculate_confidence_interval(
            self.sample,
            type='asymptotic2',
            return_estimation=True,
            cumulative=self.cumulative
        )
        
        self.left_exact, self.right_exact = calculate_confidence_interval(
            self.sample,
            type='exact',
            return_estimation=False,
            cumulative=self.cumulative
        )
    def plot_it_all(self):
        plt.figure(figsize=(14, 6))
        
        draw_confidence_interval(
            self.left_asymptotic1,
            self.right_asymptotic1,
            self.estimation1,
            ylim=(None, None),
            color_estimation='yellow',
            color_interval='#FF3300',
            label_estimation='$1/\overline{X}$',
            label_interval='Интервал Вальда 1',
            label_element='Элемент выборки'
        )
        
        draw_confidence_interval(
            self.left_asymptotic2,
            self.right_asymptotic2,
            self.estimation2,
            ylim=(None, None),
            color_estimation='#00CC66',
            color_interval='blue',
            label_estimation='$\sqrt{2/\overline{X^2}}$',
            label_interval='Интервал Вальда 2',
        )
        
        draw_confidence_interval(
            self.left_exact,
            self.right_exact,
            ylim=self.ylim,
            color_estimation='red',
            color_interval='#00CC66',
            label_estimation='$\\overline{X}}$',
            label_interval='Точный интервал',
        )
    
```

```python
def count_and_draw_proportion(
    n_samples = 100000,  # Количество выборок             
    param_value = 1,  # Истинное значение параметра       
    max_sample_size = 30,
    theta = 1,
    alpha=0.95
):

    samples = np.random.exponential(scale=1/theta, size=(n_samples, max_sample_size))  # выборка
    ir = Interval_relations(sample=samples)
    ir.count_it_all()
    def get_coverage_rate(left, right, theta):
        return np.mean(
            (left <= theta) & (theta <= right),
            axis=-1,
        )
    coverage_rate_asymptotic1 = get_coverage_rate(ir.left_asymptotic1.T, ir.right_asymptotic1.T, ir.theta)
    coverage_rate_asymptotic2 = get_coverage_rate(ir.left_asymptotic2.T, ir.right_asymptotic2.T, ir.theta)
    coverage_rate_exact = get_coverage_rate(ir.left_exact.T, ir.right_exact.T, ir.theta)
    
    sample_size_grid = np.arange(max_sample_size) + 1
        
    with sns.axes_style('darkgrid'):
        plt.figure(figsize=(14, 6))
        plt.plot(sample_size_grid, coverage_rate_asymptotic1, lw=5, label='Интервал Вальда 1')
        plt.plot(sample_size_grid, coverage_rate_asymptotic2, lw=5, label='Интервал Вальда 2')
        plt.plot(sample_size_grid, coverage_rate_exact, lw=5, label='Точный интервал')
        plt.hlines(alpha, 0, max_sample_size, lw=5, color='gray', linestyles='--', alpha=0.5)
        if theta == param_value:
            plt.ylim((0.9, 1.0))
        plt.xlim((0, max_sample_size))
        plt.xticks(np.arange(max_sample_size+1))
        plt.legend()
        plt.xlabel('Размер выборки')
        plt.ylabel('Доля покрытия')
        if theta == param_value:
            plt.title('Зависимость доли покрытия от размера выборки')
        else:
            plt.title('Зависимость доли покрытия от размера выборки при ложной $\\theta$')
```

Точный интервал оправдывает свое благородное имя - действительно ведет себя лучше всех, а вот второй интервал Вальда не слишком быстро сходится к завещанным $0.95$. Также все дружно сремятся к нулю при фейковом $\theta=2$, и это не может не радовать*.

**3.** Для всех исследуемых критериев оцените вероятность ошибки 1 рода и постройте для них доверительный интервал. Пример оценки стоит посмотреть в семинарском ноутбуке, но в данном задании требуется построить такую оценку в засимимости от размера выборки.

```python
n_samples=10000
max_sample_size=500
param_value = 1  
samples = np.random.exponential(scale=1/param_value, size=(n_samples, max_sample_size)) 
ir = Interval_relations(sample=samples, cumulative=True)
ir.count_it_all()
true = np.full((n_samples, max_sample_size), param_value)
asymptotic1_reject = ~(ir.left_asymptotic1 <= true) | ~(true <= ir.right_asymptotic1)
asymptotic2_reject = ~(ir.left_asymptotic2 <= true) | ~(true <= ir.right_asymptotic2)
exact_reject = ~(ir.left_exact <= true) | ~(true <= ir.right_exact)


asymptotic1_probability = asymptotic1_reject.cumsum(axis=-1) / (np.arange(max_sample_size) + 1)
asymptotic2_probability = asymptotic2_reject.cumsum(axis=-1) / (np.arange(max_sample_size) + 1)
exact_probability = exact_reject.cumsum(axis=-1) / (np.arange(max_sample_size) + 1)

sample_size_grid = np.arange(max_sample_size) + 1

with sns.axes_style('darkgrid'):
        plt.figure(figsize=(14, 6))
        plt.plot(sample_size_grid, asymptotic1_probability.mean(axis=0)*100, lw=4, label='Критерий 1')
        plt.plot(sample_size_grid, asymptotic2_probability.mean(axis=0)*100, lw=4, label='Критерий 2')
        plt.plot(sample_size_grid, exact_probability.mean(axis=0)*100, lw=4, label='Точный')
        plt.xlim((0, max_sample_size))
        plt.xticks(np.arange(max_sample_size+1, step=50))
        plt.legend()
        plt.xlabel('Размер выборки')
        plt.ylabel('Процент ошибок')
        plt.title('Зависимость доли ошибок от размера выборки')

```

Какие критерии не контролируют вероятность ошибки 1 рода на уровне 5\%?

*Второй критерий на порядок отстает от остальных двух по скорости приближения к $5\%$*  
*Первый критерий чуть хуже точного, но оба держатся молодцом.*

Оцените мощность критериев в зависимости от значения $\theta$ и размера выборки. Пример такой оценки есть в семинарском ноутбуке.

```python
def calculate_powers(param_value = 2, n_samples=1000, max_sample_size=500):
    samples = np.random.exponential(scale=1/param_value, size=(n_samples, max_sample_size)) 
    ir = Interval_relations(sample=samples, cumulative=False)
    ir.count_it_all()
    param = np.full(n_samples, param_value)
    true = np.full(n_samples, theta)
    asymptotic1_accept = (ir.left_asymptotic1 <= true) & (true <= ir.right_asymptotic1) 
    asymptotic2_accept = (ir.left_asymptotic2 <= true) & (true <= ir.right_asymptotic2) 
    exact_accept = (ir.left_exact <= true) & (true <= ir.right_exact) 
    asymptotic1_probability = asymptotic1_accept.sum() / n_samples
    asymptotic2_probability = asymptotic2_accept.sum() / n_samples
    exact_probability = exact_accept.sum() / n_samples
    return 1 - asymptotic1_probability, 1 - asymptotic2_probability, 1 - exact_probability

    
n_grid = [10, 50, 100, 250, 500]
p_grid = np.linspace(0.5, 2, 50)
asymptotic1_powers = np.zeros((len(n_grid), len(p_grid)))
asymptotic2_powers = np.zeros((len(n_grid), len(p_grid)))
exact_powers = np.zeros((len(n_grid), len(p_grid)))

for i, n in enumerate(n_grid):
    for j, p in enumerate(p_grid):
        pow = calculate_powers(param_value=p, n_samples=1000, max_sample_size=n)
        asymptotic1_powers[i, j], asymptotic2_powers[i, j], exact_powers[i, j]  = pow[0], pow[1], pow[2]

with sns.axes_style('darkgrid'):
    fig, (ax0, ax1, ax2) = plt.subplots(ncols=1, nrows=3, figsize=(12, 15))
    ax0.set_title("Мощность в зависимости от $n$ и param_value")
    labels = []
    for i, n in enumerate(n_grid):
        labels.append(f"n = {n}")
        ax0.plot(p_grid, asymptotic1_powers[i], lw=3)
        ax1.plot(p_grid, asymptotic2_powers[i], lw=3)
        ax2.plot(p_grid, exact_powers[i], lw=3)
    
    ax0.set_ylabel('Критерий 1, мощность')
    ax1.set_ylabel('Критерий 2, мощность')
    ax2.set_ylabel('Точный,  мощность')
    ax2.set_xlabel('param_value')
    ax0.legend(labels)
    plt.show()
```

Что можно теперь сказать про эти критерии?

*Все выглядит очень достойно. Минимум, как и должно быть, приходится на истинное значение $1$.*  
*При фиксированном $param\_value$ мощность тем больше, чем больше $n$. Логично, ведь $n$ является показателем уровня информации.  
Особой разницы между критериями в поведении не наблюдаю, у всех здесь примерно одинаковый тренд.*
